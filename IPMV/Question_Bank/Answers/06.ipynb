# Write a Python Program for HPF mask

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
!wget -q -O test1.png "https://drive.google.com/uc?export=download&id=1ADbMz3wsktYb8txPkwN5FDpk0A-nFQmH"
```

```python
image=Image.open('test1.png')
image=image.convert('L')
n,m=image.size
im1=np.array(image)
```

```python
w=3
filt=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
f=np.pad(im1,((1,1),(1,1)),'constant')
hpf=np.zeros((n,m))
r,c=f.shape
for k in range(3):
    for i in range(1,r-1):
        for j in range(1,c-1):
            window=f[(i-1):(i+2),(j-1):(j+2)]
            B=np.multiply(window,filt)
            S=(np.sum(B))/(w*w)
            hpf[i-1,j-1]=S

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(im1,cmap='gray')
plt.title("Original image",fontsize=8)
plt.subplot(2,2,2)
plt.imshow(np.uint8(hpf),cmap='gray')
plt.title('HIGH PASS FILTER',fontsize=8)
```
