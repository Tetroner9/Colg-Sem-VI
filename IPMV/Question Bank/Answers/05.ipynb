# Write a Python Program for LPF Mask.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
!wget -q -O test1.png "https://drive.google.com/uc?export=download&id=1ADbMz3wsktYb8txPkwN5FDpk0A-nFQmH"
```

```python
image=Image.open('test1.png')
image=image.convert('L')
n,m=image.size
im1=np.array(image)
```

```python
w = 5
avg = np.ones((w,w))
```

```python
f=np.pad(im1,((2,2),(2,2)),'constant')
```

```python
lpf=np.zeros((m,n))
r,c=f.shape

for i in range(2,r-2):
    for j in range(2,c-2):
        window=f[(i-2):(i+3),(j-2):(j+3)]
        B=np.multiply(window,avg)
        S=(np.sum(B))/(w*w)
        lpf[i-2,j-2]=S
plt.subplot(1,2,1)
plt.imshow(im1,cmap='gray')
plt.title("Original image",fontsize=8)
plt.subplot(1,2,2)
plt.imshow(np.uint8(lpf),cmap='gray')
plt.title("Low pass filter (blur) image",fontsize=8)
```
