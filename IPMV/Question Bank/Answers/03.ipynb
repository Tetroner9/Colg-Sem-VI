# Write a Python Program for Log Transform and Power Law Transform

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
!wget -O lena_gray.tif https://github.com/Tetroner9/Colg-Sem-VI/raw/master/IPMV/sample/lena_gray.tif
```

```python
# Fourier Transform
image = Image.open('lena_gray.tif')
im1 = np.array(image)
fourier = np.fft.fftshift(np.fft.fft2(im1))
c = 50
log = c * (np.log(abs(fourier)) + 1)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(abs(fourier), cmap='gray')
plt.title("Fourier Transform")

plt.subplot(1, 2, 2)
plt.imshow(log, cmap='gray')
plt.title("Log Transformed Fourier")
```

```python
# Image Manipulation
brighter = im1 ** 0.5
no_change = im1 ** 1
darker = im1 ** 1.5

plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.imshow(im1, cmap='gray')
plt.title("ORIGINAL IMAGE", fontsize=10)

plt.subplot(2, 2, 2)
plt.imshow(brighter, cmap='gray')
plt.title("BRIGHTER IMAGE", fontsize=10)

plt.subplot(2, 2, 3)
plt.imshow(no_change, cmap='gray')
plt.title("IDENTITY IMAGE", fontsize=10)

plt.subplot(2, 2, 4)
plt.imshow(darker, cmap='gray')
plt.title("DARKER IMAGE", fontsize=10)

plt.show()
```
