```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Read the image
image_path = "lena_colored.png"  # Replace with image name
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
def log_transform(image):
    # Applying log transform
    c = 255 / np.log(1 + np.max(image))
    log_transformed_image = c * (np.log(image + 1))
    log_transformed_image = np.uint8(log_transformed_image)
    return log_transformed_image

def power_law_transform(image, gamma):
    # Applying power law (gamma) transform
    power_law_transformed_image = np.power(image / 255.0, gamma)
    power_law_transformed_image = np.uint8(power_law_transformed_image * 255)
    return power_law_transformed_image
```

```python
# Check if the image has been successfully loaded
if image is None:
    print("Error: Could not read the image.")
else:
    # Display the original image
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Perform log transform and display
    log_transformed_image = log_transform(image)
    plt.subplot(2, 2, 2)
    plt.imshow(log_transformed_image, cmap='gray')
    plt.title('Log Transform')
    plt.axis('off')

    # Perform power law (gamma) transform and display
    gamma = 0.5  # You can adjust the value of gamma here
    power_law_transformed_image = power_law_transform(image, gamma)
    plt.subplot(2, 2, 3)
    plt.imshow(power_law_transformed_image, cmap='gray')
    plt.title('Power Law Transform (Gamma = 0.5)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
