# Write a Python Program for Bit plane Slicing and Image Negative

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
!wget -O lena_color.tif https://github.com/Tetroner9/Colg-Sem-VI/raw/master/IPMV/sample/lena_color.tif
```

```python
# Load the image and convert it to grayscale
image = Image.open('lena_color.tif')
image = image.convert('L')
wd, ht = image.size
```

```python
# Convert image to numpy array
im1 = np.array(image)

# Initialize arrays to store individual bit planes
I1 = np.zeros((ht, wd))
I2 = np.zeros((ht, wd))
I3 = np.zeros((ht, wd))
I4 = np.zeros((ht, wd))
I5 = np.zeros((ht, wd))
I6 = np.zeros((ht, wd))
I7 = np.zeros((ht, wd))
I8 = np.zeros((ht, wd))
```

```python
# Iterate through each pixel to extract individual bit planes
for i in range(ht):
    for j in range(wd):
        temp = format(im1[i][j], '08b')  # Convert pixel value to binary
        I1[i][j] = int(temp[0])  # MSB
        I2[i][j] = int(temp[1])
        I3[i][j] = int(temp[2])
        I4[i][j] = int(temp[3])
        I5[i][j] = int(temp[4])
        I6[i][j] = int(temp[5])
        I7[i][j] = int(temp[6])
        I8[i][j] = int(temp[7])  # LSB

# Plotting
plt.figure(figsize=(14, 7))

# Original Image
plt.subplot(2, 4, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# MSB Bit Image
plt.subplot(2, 4, 2)
plt.imshow(I1, cmap='gray')
plt.title("MSB Bit Image")
plt.axis('off')

# Individual Bit Images
for i in range(7):
    plt.subplot(2, 4, i+3)
    plt.imshow(eval(f'I{i+2}'), cmap='gray')
    plt.title(f"Bit {i+2}")
    plt.axis('off')

# LSB Bit Image
plt.subplot(2, 4, 8)
plt.imshow(I8, cmap='gray')
plt.title("LSB Bit Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
