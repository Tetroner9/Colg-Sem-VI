```python
def sobel_operator(size):
    # Sobel operator for edge detection
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])

    sobel_y = np.array([[-1, -2, -1],
                        [0, 0, 0],
                        [1, 2, 1]])

    return sobel_x, sobel_y

def prewitt_operator(size):
    # Prewitt operator for edge detection
    prewitt_x = np.array([[-1, 0, 1],
                          [-1, 0, 1],
                          [-1, 0, 1]])

    prewitt_y = np.array([[-1, -1, -1],
                          [0, 0, 0],
                          [1, 1, 1]])

    return prewitt_x, prewitt_y

def roberts_operator(size):
    # Roberts operator for edge detection
    roberts_x = np.array([[1, 0],
                          [0, -1]])

    roberts_y = np.array([[0, 1],
                          [-1, 0]])

    return roberts_x, roberts_y

def plot_mask(mask_x, mask_y, title):
    # Plot the edge detection masks
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(mask_x, cmap='gray')
    plt.title(f'{title} - X')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(mask_y, cmap='gray')
    plt.title(f'{title} - Y')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Define the size of the masks
mask_size = 3  # Adjust the size of the masks as needed

# Generate Sobel, Prewitt, and Roberts edge detection masks
sobel_x, sobel_y = sobel_operator(mask_size)
prewitt_x, prewitt_y = prewitt_operator(mask_size)
roberts_x, roberts_y = roberts_operator(mask_size)

# Plot the edge detection masks
plot_mask(sobel_x, sobel_y, 'Sobel')
plot_mask(prewitt_x, prewitt_y, 'Prewitt')
plot_mask(roberts_x, roberts_y, 'Roberts')
```
