```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Read the image
image_path = "lena_colored.png"  # Replace with image name
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
def generate_lpf_mask(size):
    # Generate LPF mask
    mask = np.ones((size, size), dtype=np.float32) / (size * size)
    return mask

def plot_mask(mask):
    # Plot the LPF mask
    plt.imshow(mask, cmap='gray', interpolation='nearest')
    plt.title('LPF Mask')
    plt.colorbar()
    plt.show()
```

```python
# Define the size of the LPF mask
mask_size = 5  # Adjust the size of the mask as needed

# Generate LPF mask
lpf_mask = generate_lpf_mask(mask_size)

# Plot the LPF mask
plot_mask(lpf_mask)
```
