# Write a Python Program for Opening and Closing

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Read the image
image_path = "lena_binary.png"  # Replace with image name
binary_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
def apply_opening(image, kernel_size):
    # Define the structuring element (kernel) for opening
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    # Apply opening
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    return opening

def apply_closing(image, kernel_size):
    # Define the structuring element (kernel) for closing
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    # Apply closing
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return closing
```

```python
if binary_image is None:
    print("Error: Could not read the image.")
else:
    # Binarize the image (convert to binary)
    _, binary_image = cv2.threshold(binary_image, 127, 255, cv2.THRESH_BINARY)

    # Define the kernel size for opening and closing
    kernel_size = 5

    # Apply opening and closing
    opened_image = apply_opening(binary_image, kernel_size)
    closed_image = apply_closing(binary_image, kernel_size)

    # Display the original, opened, and closed images
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(binary_image, cmap='gray')
    plt.title('Original Binary Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(opened_image, cmap='gray')
    plt.title('Opened Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(closed_image, cmap='gray')
    plt.title('Closed Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
