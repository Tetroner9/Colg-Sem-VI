```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
!wget -q -O lena_gray.tif https://github.com/Tetroner9/Colg-Sem-VI/raw/master/IPMV/sample/lena_gray.tif
!wget -q -O forest.png "https://drive.google.com/uc?export=download&id=14htmrPBkR1hVT44KL2qFehoIFs_tAP3u"
```

```python
image=Image.open('forest.png')
image=image.convert('L')
wd,ht=image.size
print(wd,ht)
im1=np.array(image)
I1=np.zeros((ht,wd))
I2=np.zeros((ht,wd))
I3=np.zeros((ht,wd))
I4=np.zeros((ht,wd))
I5=np.zeros((ht,wd))
I6=np.zeros((ht,wd))
I7=np.zeros((ht,wd))
I8=np.zeros((ht,wd))
for i in range(ht):
    for j in range(wd):
        temp=format(im1[i][j],'08b') #taking each pixel and converting to binary
        I1[i][j]=int(temp[0]) #msb
        I2[i][j]=int(temp[1])
        I3[i][j]=int(temp[2])
        I4[i][j]=int(temp[3])
        I5[i][j]=int(temp[4])
        I6[i][j]=int(temp[5])
        I7[i][j]=int(temp[6])
        I8[i][j]=int(temp[7]) #lsb
plt.subplot(2,4,1)
plt.imshow(image,cmap='gray')
plt.title("original image")
plt.subplot(2,4,2)
plt.imshow(I1,cmap='gray')
plt.title("MSB BIT IMAGE")
plt.subplot(2,4,3)
plt.imshow(I2,cmap='gray')
plt.subplot(2,4,4)
plt.imshow(I3,cmap='gray')
plt.subplot(2,4,5)
plt.imshow(I4,cmap='gray')
plt.subplot(2,4,6)
plt.imshow(I5,cmap='gray')
plt.subplot(2,4,7)
plt.imshow(I6,cmap='gray')
plt.subplot(2,4,8)
plt.imshow(I7,cmap='gray')
plt.title("LSB BIT IMAGE")
```
