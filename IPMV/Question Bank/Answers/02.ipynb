# Write a Python Program for Bit plane Slicing and Image Negative

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
!wget -O lena_color.tif https://github.com/Tetroner9/Colg-Sem-VI/raw/master/IPMV/sample/lena_color.tif
```

```python
# Read the image
image_path = "lena_color.tif"  # Replace with image name
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
def bit_plane_slice(image, plane):
    # Extract the specified bit plane
    plane_slice = (image >> plane) & 1
    # Scale the values to 0-255 range for visualization
    plane_slice *= 255
    return plane_slice.astype(np.uint8)

def image_negative(image):
    negative_image = 255 - image
    return negative_image
```

```python
# Check if the image has been successfully loaded
if image is None:
    print("Error: Could not read the image.")
else:
    # Display the original image
    plt.figure(figsize=(10, 6))
    plt.subplot(3, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Perform bit-plane slicing and display in subplots
    for i in range(8):
        plane_slice = bit_plane_slice(image, i)
        plt.subplot(3, 3, i+2)
        plt.imshow(plane_slice, cmap='gray')
        plt.title(f'Bit Plane {i}')
        plt.axis('off')

    # Perform image negative operation and display
    negative_image = image_negative(image)
    plt.subplot(3, 3, 9)
    plt.imshow(negative_image, cmap='gray')
    plt.title('Negative Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
