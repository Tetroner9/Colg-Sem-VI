# Write a Python Program for Erosion and Dilation

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
!wget -q -O morph.png "https://drive.google.com/uc?export=download&id=1jQ3BxcCoG3FqzkReFvpc4MRFvCieDXyU"
```

```python
# Read the image
image_path = "morph.png"  # Replace with image name
binary_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
def apply_erosion(image, kernel_size):
    # Define the structuring element (kernel) for erosion
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    # Apply erosion
    erosion = cv2.erode(image, kernel, iterations=1)
    return erosion

def apply_dilation(image, kernel_size):
    # Define the structuring element (kernel) for dilation
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    # Apply dilation
    dilation = cv2.dilate(image, kernel, iterations=1)
    return dilation
```

```python
# Check if the image has been successfully loaded
if binary_image is None:
    print("Error: Could not read the image.")
else:
    # Binarize the image (convert to binary)
    _, binary_image = cv2.threshold(binary_image, 127, 255, cv2.THRESH_BINARY)

    # Define the kernel size for erosion and dilation
    kernel_size = 5

    # Apply erosion and dilation
    eroded_image = apply_erosion(binary_image, kernel_size)
    dilated_image = apply_dilation(binary_image, kernel_size)

    # Display the original, eroded, and dilated images
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(binary_image, cmap='gray')
    plt.title('Original Binary Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(eroded_image, cmap='gray')
    plt.title('Eroded Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(dilated_image, cmap='gray')
    plt.title('Dilated Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
